```python
import pandas as pd
import numpy as np
%matplotlib inline
```

### Reading CSV Files

```python
data = pd.read_csv('train.csv')
```

```python
df = pd.DataFrame(data)
df
```

### Removing and Filling Nan Values

```python
df.dropna().head()  #Dropping NaN values
```

```python
df.fillna(value='FILL VALUE').tail()
```

```python
df['Age'].fillna(value=df['Age'].mean()).head() #Filling Nan Values of Age Column with the mean Age
```

### Group By

```python
by_sex = df.groupby("Sex")
by_sex.mean()
```

### Concatenate Dataframe

```python
df1 = df[['Name', 'Sex', 'Age']]
```

```python
df2 = df1.head(5)
df2
```

```python
df3 = df1.tail(5)
df3
```

```python
concat = pd.concat([df2,df3])  #Concatinating dataframes df2 and df3
concat
```

### Merge Dataframe

```python
pd.merge(df2, df2, on='Age') #Merge on the basis of common data
```
